```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final_cleaned_data.csv')
```

```python
df.columns
```

```python
df['ID'] = df.index
cols = ['ID'] + [col for col in df.columns if col != 'ID']
df = df[cols]
df
```

```python
df.columns
```

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# correlation to [119. Production Wste (8.1-8.7)]
correlation_matrix = df[numerical_cols].corr()
correlation_with_target = correlation_matrix['119. PRODUCTION WSTE (8.1-8.7)']

# sort by absolute value
correlation_with_target = correlation_with_target.abs().sort_values(ascending=False)
correlation_with_target = correlation_with_target[correlation_with_target > 0.1]
correlation_with_target
```
