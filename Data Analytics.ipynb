```python
corr = nums.corr()
corr = corr['119. PRODUCTION WSTE (8.1-8.7)'].sort_values(ascending=False)

top_10 = corr.head(10)
bottom_10 = corr.tail(10)

plt.figure(figsize=(4, 5))
sns.heatmap(top_10.to_frame(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Top 10 Features Correlated with Production Waste')
plt.show()

plt.figure(figsize=(4, 5))
sns.heatmap(bottom_10.to_frame(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Bottom 10 Features Correlated with Production Waste')
plt.show()
```

```python
nums = nums.columns
nums
```

```python
# make me top 10 skewness and kurtosis for nums in a table
skewness = []
kurtosis = []

for column in nums:
    skewness.append(stats.skew(df[column]))
    kurtosis.append(stats.kurtosis(df[column]))

skewness_df = pd.DataFrame({
    'Column': nums,
    'Skewness': skewness,
    'Kurtosis': kurtosis
})
# add new column where they skew (left or right)
skewness_df['Skew'] = skewness_df['Skewness'].apply(lambda x: 'Left' if x < 0 else 'Right')

# sort top 10 skewness
skewness_df = skewness_df.sort_values(by='Skewness', ascending=False)
display(skewness_df.head(10))
```
