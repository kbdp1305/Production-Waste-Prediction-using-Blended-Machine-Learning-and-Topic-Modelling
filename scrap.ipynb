```python
import pandas as pd
```

```python
# key_word = "production waste OR chemical waste OR toxic waste OR reason of waste OR us waste problem OR us bad waste system problem OR us environment waste problem"
key_word = "us environment waste problem"
token = ""

filename_template = "2023_month_{month}_uewp.csv"
search_template = key_word + ' until:{end_date} since:{start_date} lang:en'

# Define months and days
months = [
    {"month": "01", "start_date": "2023-01-01", "end_date": "2023-01-31"},
    {"month": "02", "start_date": "2023-02-01", "end_date": "2023-02-28"},
    {"month": "03", "start_date": "2023-03-01", "end_date": "2023-03-31"},
    {"month": "04", "start_date": "2023-04-01", "end_date": "2023-04-30"},
    {"month": "05", "start_date": "2023-05-01", "end_date": "2023-05-31"},
    {"month": "06", "start_date": "2023-06-01", "end_date": "2023-06-30"},
    {"month": "07", "start_date": "2023-07-01", "end_date": "2023-07-31"},
    {"month": "08", "start_date": "2023-08-01", "end_date": "2023-08-31"},
    {"month": "09", "start_date": "2023-09-01", "end_date": "2023-09-30"},
    {"month": "10", "start_date": "2023-10-01", "end_date": "2023-10-31"},
    {"month": "11", "start_date": "2023-11-01", "end_date": "2023-11-30"},
    {"month": "12", "start_date": "2023-12-01", "end_date": "2023-12-31"},
]

commands = []
for m in months:
    filename = filename_template.format(month=m["month"])
    search_keyword = search_template.format(
        end_date=m["end_date"], start_date=m["start_date"]
    )
    command = f'npx --yes tweet-harvest@2.6.1 -o "{filename}" -s "{search_keyword}" -l {3000} --token "{token}"'
    commands.append(command)

# Print the commands for manual execution
for cmd in commands:
    print(cmd)
print()
```

```python
import os
tweets_dir = "tweets-data"
tweets = [os.path.join(tweets_dir, file) for file in os.listdir(tweets_dir) if file.endswith('.csv')]

df = pd.concat([pd.read_csv(tweet) for tweet in tweets])
df = df[['created_at', 'full_text']].reset_index(drop=True)
df['created_at'] = pd.to_datetime(df['created_at'])
df = df.sort_values(by='created_at').reset_index(drop=True)
df
```

```python
df.drop_duplicates(subset=['full_text'], keep='first', inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv('tweets-2023.csv', index=False)
```
